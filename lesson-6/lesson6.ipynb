```python
# Kalit so`zlar`
# | Keyword   | Vazifasi                                      |
# | --------- | --------------------------------------------- |
# | `try`     | Xatolik bo'lishi mumkin bo'lgan kodni yozamiz |
# | `except`  | Agar xatolik yuz bersa, shu yerga o‘tadi      |
# | `else`    | Agar xatolik bo‘lmasa, ishlaydi               |
# | `finally` | Xatolik bo‘lsa ham bo‘lmasa ham ishlaydi      |
```

```python
try:
    num = int(input('Biror son kriting: '))
    print(10/num)
except ZeroDivisionError:
    print('0 ga bo`lish mumkin emas')

except ValueError:
    print('faqar son kriting')
```

```python
try:
    num = int(input('Biror son kiriting: '))
    result = 10/num

except ZeroDivisionError:
    print('0 ga bo`lish mumkin emas')

except ValueError: 
    print('Harf kiritish mumkin emas raqam kriting')
else:
    print(f'Natija: {result}')

finally:
    print('Dastur tugadi')
```

```python
urinish = 0
while urinish<3:
    try:
        son = int(input('Biror son kriting: '))
        print('Siz kiritgan son:', son)
        break
    except ValueError:
        urinish+=1
        print(f'Siz xato kritdingiz, qolgan urinishlar {3-urinish}')
if urinish==3:
    print('Ko`p xato kirtildi shuning uchun dastur to`xtatildi')
```

```python
for urinish in range(3):
    try:
        son = int(input("Iltimos, biror butun son kiriting: "))
        print("✅ Siz kiritgan son:", son)
        break  # to‘g‘ri kiritilsa, sikldan chiqamiz
    except ValueError:
        print(f"❌ Noto‘g‘ri kiritdingiz! Qolgan urinishlar: {2 - urinish}")
else:
    print("❗ 3 marta noto‘g‘ri kiritdingiz. Dastur to‘xtadi.")
```

```python
# Topshiriq: Agar foydalanuvchi noto‘g‘ri indeks kiritsa, 
# unga "Mevalar soni:" degan ko‘rsatkichni chiqaring (ya'ni len(mevalar)).
mevalar = ["olma", "anor", "banan", "shaftoli", "nok"]

print("Ro'yxatdagi mevalar soni:", len(mevalar))  # avvaldan ko‘rsatamiz

try:
    i = int(input("Index kiriting (0 dan boshlanadi): "))
    print("Siz tanlagan meva:", mevalar[i])
except IndexError:
    print("❌ Bunday indeks yo‘q! Ro'yxatda faqat", len(mevalar), "ta element bor.")
except ValueError:
    print("❌ Iltimos, faqat butun son kiriting.")
```
